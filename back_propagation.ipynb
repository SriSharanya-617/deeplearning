disease propagation
2inputs 3 hidden nuurons 1 output =1-high risk output ,0-lowrisk output

```python
import numpy as np
np.random.seed(0)
X=np.array([
  [120,85,150,95], #Glucode
  [25,30,35,22]
# BMI
])
#Normalize Values
X=X/200
#Output : 1=High Risk, 0=Low Risk
Y=np.array([[1,0,0,1]])
print("X Shape:",X.shape)
print("Y Shape:",Y.shape)
```

sigmoid function

```python
def sigmoid(z):
  return 1/(1+np.exp(-z))
```

intialize parameters

```python
#Hidden Layer(3 neurons,2 inputs)
w1=np.random.randn(3,2)*0.1
b1=np.zeros((3,1))
#Output Layer(1 neuron,3 hidden inputs)
w2=np.random.randn(1,3)*0.1
b2=np.zeros((1,1))
```

forward propagation

```python
#Hidden Layer
z1=np.dot(w1,X)+b1
a1=sigmoid(z1)

#Output Layer
z2=np.dot(w2,a1)+b2
a2=sigmoid(z2)
print("Predicted Output:")
print(a2)
```

loss function

```python
m=Y.shape[1]
loss=-np.sum(Y*np.log(a2)+(1-Y)*np.log(1-a2))/m
print("Loss:",loss)
```

back propagation output layer gradient





```python
dZ2=a2-Y
dW2=(1/m)*np.dot(dZ2,a1.T)
db2=(1/m)*np.sum(dZ2,axis=1,keepdims=True)
```

hidden layer gradient
sigmoid derivative

```python
dA1=np.dot(w2.T,dZ2)
dZ1=dA1*a1*(1-a1)

dW1=(1/m)*np.dot(dZ1,X.T)
db1=(1/m)*np.sum(dZ1,axis=1,keepdims=True)
```

update parameters

```python
learning_rate = 0.5

# Update weights and biases
w2 = w2 - learning_rate * dW2
b2 = b2 - learning_rate * db2

w1 = w1 - learning_rate * dW1
b1 = b1 - learning_rate * db1
```

complete training loop

```python
epoch=5000
for i in range(epoch):
#Forward
  Z1=np.dot(w1,X)+b1
  A1=sigmoid(Z1)

  Z2=np.dot(w2,A1)+b2
  A2=sigmoid(Z2)

#loss
  loss =- np.sum(Y*np.log(A2+1e-8)+(1-Y)*np.log(1-A2+1e-8))/m

#Backward
  dZ2=A2-Y
  dW2=(1/m)*np.dot(dZ2,A1.T)
  db2=(1/m)*np.sum(dZ2,axis=1,keepdims=True)
  dA1=np.dot(w2.T,dZ2)
  dZ1=dA1*A1*(1-A1)

  dW1=(1/m)*np.dot(dZ1,X.T) # Changed dw1 to dW1 for consistency
  db1=(1/m)*np.sum(dZ1,axis=1,keepdims=True)

#Update
  w2 -= 0.5*dW2
  b2 -= 0.5*db2
  w1 -= 0.5*dW1 # Removed ' I' and changed dw1 to dW1
  b1 -= 0.5*db1

  if i%100 == 0:
    print("Epoch:",i,"Loss:",loss)
```
